```python
myfile = open('/content/ai_updates.txt')
mytext = myfile.read()
myfile.close()
print(mytext)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
mytokenizer = Tokenizer()
mytokenizer.fit_on_texts([mytext])
```

```python
word_index = mytokenizer.word_index
word_index
```

```python
total_words = len(word_index) + 1
print("Vocabulary size:" , total_words)
```

```python
lines = mytext.split('\n')
print(lines[0])
print()
print(lines[1])
print()
print(lines[2])
```

* For training language models - generate n-gram sequences
* for each line in the text, we create sequences that incrementally add one word at a time, which will train model to predict the next word given a sequence.

```python
my_input_sequences = []

for line in lines:
  token_list = mytokenizer.texts_to_sequences([line])[0]
  #print(token_list)
  for i in range(1, len(token_list)):
    n_gram_sequence = token_list[ :i+1]
    #print(n_gram_sequence)
    my_input_sequences.append(n_gram_sequence)
```

```python
print(my_input_sequences)
```

```python
print(my_input_sequences[0])
print(my_input_sequences[1])
print(my_input_sequences[2])
```

**Padding**

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_seq_len = max([len(seq) for seq in my_input_sequences])
my_input_sequences = pad_sequences(my_input_sequences , maxlen=max_seq_len , padding='pre')
print(my_input_sequences)
```

```python
import numpy as np

X = my_input_sequences[:,:-1]
y = my_input_sequences[:,-1]
```

```python
X
```

```python
X.shape
```

```python
y = y.reshape(-1,1)
y.shape
```

```python
y
```

```python
from tensorflow.keras.utils import to_categorical

yt = to_categorical(y , num_classes=total_words)
yt
```

```python
print("Feature shape:" , X.shape)
print("Feature shape:" , yt.shape)
```

```python
from tensorflow.keras import models, layers

model = models.Sequential()
model.add(layers.Embedding(total_words , output_dim=50 , input_length=max_seq_len-1))
model.add(layers.Flatten()) # Add a Flatten layer here
model.add(layers.Dense(16 , activation='relu'))
model.add(layers.Dense(288 , activation='softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
model.fit(X , yt , epochs = 100)
```
